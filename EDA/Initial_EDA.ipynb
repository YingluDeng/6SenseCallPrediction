```python
calls['date'] =  pd.to_datetime(calls['date'])
calls['day'] = [e.strftime('%a') for e in calls['date']]
calls['month'] = calls['date'].dt.strftime('%b')

fig = plt.figure(figsize=(10,5))
# calls.groupby(['call_disposition', 'day']).size().to_frame(name='counts').unstack('day').plot.bar()

grouped_calls = calls[calls["day"].isin(['Wed', 'Thu', 'Fri', 'Mon', 'Tue'])].groupby(['call_disposition', 'day']).size().to_frame(name='count')
calls_ds = list(set(calls['call_disposition']))
days = list(set(calls['day']))

# for i in calls_ds:
#     for j in days:
#         if j not in grouped_calls.loc[i].index:
#             grouped_calls.loc[i, "j"] = [0]

events['date'] =  pd.to_datetime(events['date'])
events['day'] = [e.strftime('%a') for e in events['date']]
events['month'] = events['date'].dt.strftime('%b')
grouped_events = events[events["day"].isin(['Wed', 'Thu', 'Fri', 'Mon', 'Tue'])].groupby(['activity_type', 'day']).size().to_frame(name='count')
fig, ax = plt.subplots(figsize=(15,7))
grouped_events.unstack('day').plot.bar(ax=ax)
plt.title("Activity Count by Day of Week ")
plt.xlabel("Activity Type")
plt.ylabel("Count")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,7))
grouped_events.loc['email_open'].loc[['Mon', 'Tue', 'Wed', 'Thu', 'Fri'], :].plot.bar(ax=ax, color='purple')
plt.title("Number of Email Open by Day of Week ")
plt.xlabel("Day of Week")
plt.ylabel("Email Open Count")
plt.show() 
fig, ax = plt.subplots(figsize=(15,7))
grouped_events.loc['email_click'].loc[['Mon', 'Tue', 'Wed', 'Thu', 'Fri'], :].plot.bar(ax=ax, color='orange')
plt.title("Number of Email Click by Day of Week ")
plt.xlabel("Day of Week")
plt.ylabel("Email Click Count")
plt.show() 
fig, ax = plt.subplots(figsize=(15,7))
grouped_events.loc['form_fill'].loc[['Mon', 'Tue', 'Wed', 'Thu', 'Fri'], :].plot.bar(ax=ax, color='red')
plt.title("Number of Form Filled by Day of Week ")
plt.xlabel("Day of Week")
plt.ylabel("Form Filled Count")
plt.show() 
fig, ax = plt.subplots(figsize=(15,7))
grouped_events.loc['known_web_visit'].loc[['Mon', 'Tue', 'Wed', 'Thu', 'Fri'], :].plot.bar(ax=ax, color='green')
plt.title("Number of Website Visits by Day of Week ")
plt.xlabel("Day of Week")
plt.ylabel("Website Visits Count")
plt.show() 
fig, ax = plt.subplots(figsize=(15,7))
grouped_events.loc['attended_event'].loc[['Mon', 'Tue', 'Wed', 'Thu', 'Fri'], :].plot.bar(ax=ax, color='blue')
plt.title("Number of Attended Event by Day of Week ")
plt.xlabel("Day of Week")
plt.ylabel("Attended Event Count")
plt.show() 
```
